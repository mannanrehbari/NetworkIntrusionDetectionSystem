```python
import pandas as pd
import numpy as np
import gensim.downloader as api
```

```python
df_kdd = pd.read_pickle("./data/KDDCUP_EDA.pkl")
df_kdd.head()
```

```python
embeddingModel = api.load('glove-twitter-25')
def get_embedding_mean(word):
    try:
        return np.mean(embeddingModel[word])
    except:
        return 0.0
```

```python
df_kdd['protocol_type'] = df_kdd['protocol_type'].apply(get_embedding_mean)
df_kdd['service'] = df_kdd['service'].apply(get_embedding_mean)
df_kdd['flag'] = df_kdd['flag'].apply(get_embedding_mean)
df_kdd['target'] = df_kdd['target'].apply(get_embedding_mean)

df_kdd.head()
```

```python
df_kdd = df_kdd.drop(['target'], axis=1)
df_kdd['Attack Type'] = (df_kdd['Attack Type'] != 'normal').astype(int)
df_kdd.to_pickle("./data/KDDCUP_Preprocessed.pkl")
```
